```python
df = pd.read_csv('Churn.csv')
```

```python
df
```

```python
X = pd.get_dummies(df.drop(["Churn","Customer ID"],axis=1))
y = df["Churn"].apply(lambda x:1 if x=="Yes" else 0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
y_train.head()
```

## 2.Import Dependencies

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

## 3.Build and Compile Model

```python
# instanciating sequential class
model = Sequential()

# Adding hidden layers
model.add(Dense(units=32,activation="relu",input_dim=len(X_train.columns))) # hidden layer 1
# This is a fully connected dense layer
# units -> no of neurons in the layer (32 in this case)
# activation -> activation function (they are used to introduce non-linearity allowing the network to learn complex patterns
# and approximate more complex function than linear model)
# input_dim -> dimension of input (number of features)

model.add(Dense(units=64,activation="relu")) # hidden layer 2

model.add(Dense(units=1, activation="sigmoid")) # output layer
```

```python
# compile model

model.compile(loss="binary_crossentropy", optimizer="sgd", metrics=["accuracy"])
# loss -> a loss function measures how good a neural network model is in performing a certain task.
# optimizer -> optimizers are algorithms used to adjust a neural network's parameters (weights and biases)
# during training to minimize the loss function.
# metrics -> evaluate the performance of a model, providing insights into its accuracy, efficiency, and overall effectiveness.
```

## 4.Fit, Predict and evaluate

```python
model.fit(X_train,y_train,epochs=20,batch_size=32)

# epochs -> no of complete pass of the entire training dataset through a neural network
# batch_size -> the number of training samples processed during each forward and backward pass through a neural network
```

```python
# predict
y_hat = model.predict(X_test)
y_hat = [0 if val < 0.5 else 1 for val in y_hat]
```

```python
y_hat
```

```python
# checking accuracy
accuracy_score(y_test, y_hat)
```

## 5. Saving and Reloading

```python
# Saving the model
model.save("basic_tensor_flow_model.keras")
```
