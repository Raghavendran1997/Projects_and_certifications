## UNDERSTANDING THE DATA

```python
data_columns = {"Temperature":"The temperature of the star", 
                "L":"Relative Luminosity",
               "R" : "Relative Radius",
               "A_M" : "Absolute Magnitude",
               "Color" : "General Color of Spectrum",
               "Spectral_Class" : "represent a temperature sequence, from hotter (O stars) to cooler (M stars)"}
for a,b in data_columns.items():
    print(a, ":" ,b)
```

## RELATION BETWEEN TEMPERATURE AND LUMINOSITY

```python
sns.lmplot(data, x="Temperature", y="L")
```

- temperature and luminosity does not have a linear relationship.

## RELATION BETWEEN LUMINOSITY AND RADIUS

```python
sns.lmplot(data, x="R", y="L")
```

- There is no significant realtion between luminosity and radius.

## REALATION BETWEEN TEMPERATURE AND RADIUS

```python
sns.lmplot(data,x="Temperature", y = "R")
```

- There is no significant relation between temperature and radius of the moon.

## RELATION BETWEEN SPECTRAL CLASS AND TEMPERATURE

```python
sns.catplot(data, x = "Spectral_Class", y = "Temperature")
```

- The spectral A and K have temperature around 5000 degree clesius and other are spread aro

## RELATION BETWEEN SPECTRAL CLASS AND RADIUS

```python
sns.catplot(data,x="Spectral_Class",y="R")
```

- All stars have avg radius and outliers are present.

## RELATION BETWEEN SPECTRAL CLASS AND LUMINOCITY

```python
sns.catplot(data,x="Spectral_Class", y="L")
```

- All stars similar luminosity and outliers are present.

## RELATION BETWEEN SPECTRAL CLASS AND ABSOLUTE MAGNITUDE

```python
sns.catplot(data,x="Spectral_Class",y="A_M")
```

- spectral class M and B two groups with magnitude high and low
- spectral clsass O has low magnitude.
- rest classes are spread randomly.
