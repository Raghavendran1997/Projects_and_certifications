# GOOGLE SEARCH ANALYSIS

- Approximately 3.5 billion searches are performed on Google daily, which means that approximately 40,000 searches are performed every second on Google. So Google search is a great use case for analyzing data based on search queries.


```python
import pandas as pd
import lxml
import requests
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=["manchester united", "bayern munich"])
data = data.interest_over_time()
data = data.sort_values(by=["manchester united", "bayern munich"], ascending=False)
```

```python
data.head()
```

```python
data["manchester united"].plot()
plt.title('Total Google Searches for manchester united', 
          fontweight='bold')
plt.xlabel('Year')
plt.ylabel("total count")
plt.show()
```

```python
data["bayern munich"].plot()

plt.title('Total Google Searches for bayern munich', 
          fontweight='bold')
plt.xlabel('Year')
plt.ylabel("total count")
plt.show()
```
